# **6. Data Pre-Processing**
---

- This section is emphasised on performing data manipulation over unstructured data for further processing and analysis.

- To modify unstructured data to strucuted data you need to verify and manipulate the integrity of the data by:
  - Handling missing data,

  - Handling redundant data,

  - Handling inconsistent data,

  - Handling outliers,

  - Handling typos

```python
clean_data = data.copy()
```

Handling Missing Data
    


```python
clean_data.columns                                                            #To view column name
```

```python
clean_data['gender'].unique()                                               #to find unique value in gender column
```

```python
clean_data['gender'] = clean_data['gender'].fillna(value = 'Unknown')   #missing value of gender replace with unknown and check

clean_data['gender'].unique() 
```

```python
# missing value of tenure replace with mean of same column so mean could be same for whole.

clean_data['tenure'] = clean_data['tenure'].fillna(clean_data['tenure'].mean()) 
clean_data['tenure'].isna().sum()
```

Handling Redundant Data


```python
clean_data.duplicated().any()                               #check if any duplicate exist
```

Handling Inconsistent Data

```python
clean_data['tenure'] = clean_data['tenure'].round()          #take round value of tenure column as it was in decimal form
```

Heading Outliers


```python
clean_data.describe()                     #No outlier found.                       
```

```python
clean_data.info()                                       #No outlier found.   
```

Handling Typos


```python
for x in clean_data.columns:
    print(clean_data[x].unique())
    print('\n, --------------------------------------------------------------------------')
    
```

# **8. Exploratory Data Analysis**
---

- This section is emphasised on asking the right questions and perform analysis using the data.

- Note that there is no limit how deep you can go, but make sure not to get distracted from right track.

```python
#Trend of young users on Facebook


figure, ax = plt.subplots(figsize=[18,12])

year_wise_evolution = clean_data.groupby('dob_year').agg({'userid':'count'})
year_wise_evolution.plot(kind='line', color='maroon', ax=ax, marker='o' )


# background_color = 'midnightblue'
# figure.set_facecolor(background_color)


plt.xticks(rotation=70, size=14)

plt.xlabel("YEARS", size=14)

plt.ylabel("Users Counts", size=14)

plt.title("Year wise Trend of users on Facebook", size=20)

plt.grid(linestyle='--')

plt.tight_layout()
```

```python
#male or female friend count.

clean_data['gender'].value_counts().plot(kind='pie', autopct='%3.1f%%',
                                         shadow=True, startangle=100,
                                         cmap='cool', legend=True, explode=[0.1,0,0])



plt.title("Male and Female Active on Facebook", size=14)

plt.grid(linestyle=':')

plt.tight_layout()
```

```python
agewise_users = clean_data.groupby('age')['userid'].count().sort_values(ascending=True)

# Define a colormap for the bars (automatically generates colors)
colormap = plt.cm.get_cmap('tab20', len(agewise_users))

figure, ax = plt.subplots(figsize=[15, 15])

# Plot the horizontal bar chart with colors from the colormap
agewise_users.plot(kind='barh', ax=ax, color=colormap(range(len(agewise_users))))

# Customize the chart appearance
plt.title("Most Active Age Group on Facebook", size=16)
plt.xlabel("Number of Users", size=14)
plt.ylabel("Age Groups", size=14)

plt.grid(linestyle='--')
plt.tight_layout()
```

```python
#Comparison between mobile and web usage by user.

figure, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=[15, 7])

#for ax1
user_do = clean_data[['mobile_likes', 'www_likes']].sum()
user_do.plot(kind='bar', ax=ax1, stacked=True, color=['orange', 'green'])

#set labels and titles
ax1.set_xlabel('Likes does', size=14)
ax1.set_ylabel('sum of Likes does (in mn)', size=14)
ax1.set_title('Mobile like does Vs www like does', size=14)
ax1.grid(linestyle=':')



#for ax2
user_receive = clean_data[['mobile_likes_received', 'www_likes_received']].sum()
user_receive.plot(kind='bar', ax=ax2, stacked=True, color=['purple', 'gold'])

ax2.set_xlabel('Likes Received', size=14)
ax2.set_ylabel('sum of Likes Received (in mn)', size=14)
ax2.set_title('Mobile like Received Vs www like Received', size=14)
ax2.grid(linestyle=':')





plt.tight_layout()
```

```python
#elders more prefer surfing then making friends(revenue generation)


figure, ax = plt.subplots(figsize=[10, 7])

#bins = np.arange(0, 3500, 500)

new_tenure = clean_data.groupby('tenure').agg({'friend_count':'sum'})
new_tenure.plot(kind='hist', color='olive', ax=ax)

plt.xlabel('Tenure', size=12)
plt.ylabel('Sum of friend_count', size=12)
plt.title('Peoples Connection trend with time', size=14)

plt.grid(linestyle=':')
plt.tight_layout()
```

```python
clean_data.head()
```

```python
#People like to connect out of box(desire of more connection).

figure, ax = plt.subplots(figsize=[15, 7])
compare = clean_data.groupby('gender').agg({'friend_count':'sum', 'friendships_initiated':'sum', 'likes':'sum', 'likes_received':'sum'})

compare.plot(kind='bar', stacked=True, ax=ax)

plt.xlabel('gender', size=14)
plt.ylabel('Sum of friends and Likes', size=14)
plt.title('Out of the box connection on Facebook', size=16)

plt.grid(linestyle=':')
plt.tight_layout()
```
